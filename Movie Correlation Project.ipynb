```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjust the config of the plots

#use this otherwise change of datatypes won't work
pd.options.mode.chained_assignment = None  # default='warn'

#read the data
df = pd.read_csv(r'D:\DATA ANALYST\PROJECTS\PYTHON\Beginner- Corelation Proj\movies.csv')

#look at the data 
#top 5 rows .head()
df.head()
```

```python
#checking for missing values

for col in df.columns:
    missing_pct = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_pct))
```

```python
#dropping missing values
df = df.dropna()
```

```python
for col in df.columns:
    missing_pct = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_pct))
```

```python
#checking datatypes

df.dtypes
```

```python
#changing the datatpes

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
#correct release year column
 
df['yearcorrect'] = df['released'].astype(str).str.split().str[-3]
df.head()
```

```python
#sorting the data according to gross
df = df.sort_values(by =['gross'], inplace=False, ascending=False)
```

```python
#to see the entire data

pd.set_option('display.max_rows', None)
```

```python
#dropping the duplicates if any
df_drop_dupli=df.drop_duplicates()
df_drop_dupli.head()
```

```python
#Budget and company might have high correlation
```

```python
#scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earning')
plt.show()
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"black"}, line_kws={"color":"red"})
```

```python
#check for correlation

df.corr(method='pearson') #types of correlation: 1)pearson 2)kendall 3)spearman
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot= True)
plt.title('Correlation Matrix for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#let's look at company
df.head()
```

```python
#converting the company data type to numeric to get the correlation

df_numerize=df

for col in df_numerize.columns:
    if(df_numerize[col].dtype == 'object'):
        df_numerize[col] = df_numerize[col].astype('category')
        df_numerize[col] = df_numerize[col].cat.codes
        
df_numerize.head()
```

```python
correlation_matrix = df_numerize.corr(method='pearson')
sns.heatmap(correlation_matrix, annot= True)
plt.title('Correlation Matrix for numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Unstacking: looking at numeric data

correlation_mat =df_numerize.corr()
correlation_pairs = correlation_mat.unstack()
correlation_pairs.head()
```

```python
sorted_pairs = correlation_pairs.sort_values()
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr.head()
```
